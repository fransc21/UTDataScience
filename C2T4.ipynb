```python
# I would like to know what columns dropped
safety = safety.drop_duplicates()
safety.info()
```

```python
# re naming columns
safety = safety.rename(columns = {'Safety Observation Card Count':'SOCC'})
safety = safety.rename(columns = {'Near Miss (NM)':'NM'})
safety = safety.rename(columns = {'Environmental (ENV) Incident Count':'ENV'})
safety = safety.rename(columns = {'Property/ Equipment Damage (EQP DMG) Incident Count':'EQP'})
safety = safety.rename(columns = {'First Aid Cases (FAC)':'FAC'})
safety = safety.rename(columns = {'Medical Treatment Only (MTO)':'MTO'})
safety = safety.rename(columns = {'Restricted Work Case (RWC)':'RWC'})
safety = safety.rename(columns = {'Lost Time Incident (LTI)':'LTI'})
safety = safety.rename(columns = {'Fatality (FTL)':'FTL'})
safety = safety.rename(columns = {'Total Recordable Cases (TTL REC)':'REC'})
safety = safety.rename(columns = {'Total Class 1 Incidents':'CL1'})
safety = safety.rename(columns = {'Class 1 Incident Rate (C1IR)':'C1IR'})
safety = safety.rename(columns = {'Total Recordable Incident Rate (TRIR)':'TRIR'})
safety = safety.rename(columns = {'Lost Time Incident Rate (LTIR)':'LTIR'})
safety = safety.rename(columns = {'Total Hurt Rate (THR)':'THR'})
safety = safety.rename(columns = {'Count of STARs':'STARC'})
```

```python
# replacing nans
safety['Well Customer'] = safety['Well Customer'].fillna(0)
safety['C1IR'] = safety['C1IR'].fillna(0)
safety['TRIR'] = safety['TRIR'].fillna(0)
safety['LTIR'] = safety['LTIR'].fillna(0)
safety['THR'] = safety['THR'].fillna(0)
```
